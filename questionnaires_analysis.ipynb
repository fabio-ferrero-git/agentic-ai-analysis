#### Import

```python
from analysis_functions import utils as u, statistic as s
from analysis_functions import pre_processing
from itertools import product
import numpy as np
```

#### Get preprocessed data

```python
preprocessed_data = pre_processing.pre_process()
study_questions = pre_processing.study_questions()

questionnaires_results = preprocessed_data.get('questionnaires_results')
study_sample = preprocessed_data.get('study_sample')
```

## Kruskal Wallis

```python
for quest, anova_dfs in s.kruskal_wallis(questionnaires_results, study_questions).items():
    print(f'Questionnaire: {quest}')
    print(f'Type: {"PRE TASK" if quest in u.pre_task_names else "POST TASK"}')
    display(anova_dfs)
```

---
## Mann Whitney

```python
for quest, t_test_dfs in s.mann_whitney(questionnaires_results, study_questions).items():
    print(f'Questionnaire: {quest}')
    print(f'Type: {"PRE TASK" if quest in u.pre_task_names else "POST TASK"}')
    display(t_test_dfs)
```

## Equivalence Test (TOAST)

```python
for quest, eq_dfs in s.equivalence_testing(questionnaires_results, study_questions).items():
    print(quest)
    display(eq_dfs)
```

### Compute questionnaires aggregations

```python
pre_task_aggregation = {
    sys: {q_name: u.get_aggregation_score(questionnaires_results[sys][q_name], q_name) 
          for q_name in u.pre_task_names} 
    for sys in questionnaires_results.keys()
}
pre_task_aggregation_pairs = list(product(pre_task_aggregation.keys(), pre_task_aggregation[0].keys()))


tia_post_task_aggregation = {
    sys: u.tia_post_task_aggregation_score(questionnaires_results[sys]['tia_post_task'])
    for sys in questionnaires_results
}


tlx_aggregation = {
    sys: np.array(questionnaires_results[sys]['tlx_questionnaire']).transpose().tolist()
    for sys in questionnaires_results
}
# Fourth question should be inverted
for i in range(3):
    tlx_aggregation[i][3] = [100 - x for x in tlx_aggregation[i][3]]


grid_questionnaire_aggregation = {
    sys: np.array(questionnaires_results[sys].get('grid_questionnaire')).transpose().tolist()  
    for sys in questionnaires_results if questionnaires_results[sys].get('grid_questionnaire') is not None
}

tlx_aggregation_unique = {
    sys: np.mean(questionnaires_results[sys]['tlx_questionnaire'], axis=1)
    for sys in questionnaires_results
}
```

## Equivalence Test (TOAST) and Mann Whitney

```python
df1, df2 = s.np_test_toast_post_task_aggregation(tia_post_task_aggregation)
display(df1)
display(df2)
```

## NASA-TLX Aggregate Analysis

The analysis below examines the aggregated NASA-TLX scores (mean of all TLX questions per participant) across systems.
This provides an overall workload measure in addition to the individual dimension analysis above.
Calculate and display the aggregated NASA-TLX results

```python
import matplotlib.pyplot as plt
tlx_anova_df, tlx_mw_df, tlx_tost_df = s.analyze_tlx_aggregate(tlx_aggregation_unique)

print("Kruskal-Wallis Test (Non-parametric ANOVA) for NASA-TLX Aggregate:")
display(tlx_anova_df)

print("\nMann-Whitney U Tests (Non-parametric t-tests) for NASA-TLX Aggregate:")
display(tlx_mw_df)

print("\nTOST Equivalence Tests for NASA-TLX Aggregate:")
display(tlx_tost_df)

# Visualize the aggregated NASA-TLX scores
plt.figure(figsize=(10, 6))
boxplot_data = [tlx_aggregation_unique[sys] for sys in sorted(tlx_aggregation_unique.keys())]
box = plt.boxplot(boxplot_data, patch_artist=True,
                  labels=[f'System {i}' for i in sorted(tlx_aggregation_unique.keys())])

# Color the boxes
colors = ['lightblue', 'lightgreen', 'lightpink']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.title('NASA-TLX Aggregate Scores by System')
plt.ylabel('NASA-TLX Score (Lower is Better)')
plt.grid(True, linestyle='--', alpha=0.7)

# Add means as text
for i, sys in enumerate(sorted(tlx_aggregation_unique.keys())):
    mean_val = np.mean(tlx_aggregation_unique[sys])
    plt.text(i + 1, np.min(tlx_aggregation_unique[sys]), f'Mean: {mean_val:.2f}',
             horizontalalignment='center', size='small', color='blue', weight='semibold')

plt.tight_layout()
plt.show()
```
